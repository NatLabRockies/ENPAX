# Hybrid Solar+BESS and Standalone BESS

```python
import os
import sys
from pprint import pprint

# 1. Get the absolute path of the folder containing THIS script
current_dir = os.getcwd() # or os.path.dirname(os.path.abspath(__file__)) if in a .py file

# 2. Find the 'GenStorBOSSE' root directory
# This looks for the folder that actually contains the 'genstor' package
project_root = current_dir
while project_root != os.path.dirname(project_root): # stop at drive root
    if os.path.exists(os.path.join(project_root, 'genstor')):
        break
    project_root = os.path.dirname(project_root)

# 3. Add that root to sys.path
if project_root not in sys.path:
    sys.path.append(project_root)

# 4. Now the import will work
from genstor.runner import CentralRunner

# 5. Correctly locate your config file relative to that root
config_path = os.path.join(project_root, "configs", "solar_bess_plus_standalone_bess_hybrid.yaml")

result = CentralRunner.from_yaml(config_path).run()
display(result.summary())
```

```python
pprint(result)
```
